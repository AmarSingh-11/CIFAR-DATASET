```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        data = pickle.load(fo, encoding='bytes')
    return data
```

```python
import pickle
import numpy as np
import pandas as pd

# Initialize empty lists to store images, labels, and batch information
all_images = []
all_labels = []
all_batches = []  # Add this list to store batch information

# Define a list of filenames to unpickle
filenames = ['data_batch_1', 'data_batch_2', 'data_batch_3', 'data_batch_4', 'data_batch_5','test_batch']  # Add more filenames as needed

# Loop through the filenames and unpickle each file
for filename in filenames:
    with open(filename, 'rb') as file:
        data_batch = pickle.load(file, encoding='bytes')
        
        # Extract images, labels, and batch information from the data batch
        images = data_batch[b'data']
        labels = data_batch[b'labels']
        
        # Append the images, labels, and batch information to their respective lists
        all_images.append(images)
        all_labels.extend(labels)
        all_batches.extend([filename] * len(images))  # Store batch information for each image

# Concatenate all the images into a single NumPy array
all_images = np.concatenate(all_images, axis=0)

# Reshape the image data to have one column per pixel channel (RGB)
all_images = all_images.reshape(-1, 3, 32, 32).transpose(0, 2, 3, 1)  # Assuming image dimensions are 32x32x3 (HxWxC)

# Flatten the image arrays to store them as pixel values
all_images = all_images.reshape(-1, 32 * 32 * 3)

# Create a DataFrame to store the data
data = pd.DataFrame()

# Add columns for pixel values, labels, and batch information
for i in range(32 * 32 * 3):
    data[f'Pixel_{i + 1}'] = all_images[:, i]

data['Label'] = all_labels
data['Batch'] = all_batches  # Add batch information column

# Now, 'data' contains columns for pixel values, labels, and batch information
# You can save it to a CSV file using data.to_csv('your_filename.csv', index=False)
```

```python
# Iterate through the first image of each data batch
for i in range(len(filenames)):
    first_image = all_images[i]  # Get the first image from the ith data batch
    
    # Print the shape and dimensions of the first image
    print(f"Shape of the first image in {filenames[i]}: {first_image.shape}")
    print(f"Dimensions of the first image in {filenames[i]}: {first_image.ndim}")
```
